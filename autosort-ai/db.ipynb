```python
from datetime import datetime
```

```python
db = {
    "users": {}
}
```

```python
def generate_user_id():
    user_count = len(db["users"]) + 1
    return f"{user_count:08}"
```

```python
def init_user(user_id=None):
    if user_id is None:
        user_id = generate_user_id()
    if user_id not in db["users"]:
        db["users"][user_id] = {
            "history": []
        }
    return user_id
```

```python
def save_interaction(user_id, query, intent, response):
    user_data = init_user(user_id)
    db["users"][user_id]["history"].append({
        "query": query,
        "intent": intent,
        "response": response,
        "timestamp": datetime.utcnow().isoformat()
    })
```

```python
def get_user_history(user_id, limit=None):
    if user_id in db["users"]:
        history = db["users"][user_id]["history"]
        return history[-limit:] if limit else history
    return []
```
